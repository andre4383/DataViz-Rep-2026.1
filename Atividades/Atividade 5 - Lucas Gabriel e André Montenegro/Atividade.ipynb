# Conjunto de dados utilizados
## Pesquisa Nacional por Amostra de Domicílios Contínua Trimestral.

- Pessoas de 14 anos ou mais de idade, total, na força de trabalho, ocupadas, desocupadas, fora da força de trabalho, em situação de informalidade e respectivas taxas e níveis, por sexo.
- **Taxa de desocupação**, na semana de referência, das pessoas de 14 anos ou mais de idade.
- Pesquisa realizada desde o **1º Trimestre de 2012 até 4º Trimestre de 2025**.
- Sexo: Homens e Mulheres.
- Pesquisa realizada apenas no estado de Pernambuco.

-----
## Dicionário

### Significado das Colunas

- **periodo**: Representa a referência temporal dos dados.
    - As datas (ex: 2012-01-01, 2012-04-01) indicam o início de cada trimestre.
    - `2012-01-01`: 1º Trimestre de 2012.
    - `2012-04-01`: 2º Trimestre de 2012, e assim por diante.

- **total**: É o índice geral para toda a população analisada naquele período (provavelmente a taxa de desocupação ou desemprego).

- **homens**: O índice específico para a população masculina.

- **mulheres**: O índice específico para a população feminina.


```python
import pandas as pd

url = "https://docs.google.com/spreadsheets/d/1BRIthYFiT3i39WGEk3AQd24BH8rT4EokJ4WsEiMp1jE/export?format=csv"
data = pd.read_csv(url)
data
```

```python
data.info()
```

## Médias bienal

```python
data = data[data['total'] != 0]
```

```python
anos_remover = [2020, 2021, 2022, 2026]
```

```python
data_filtrada = data[~data['periodo'].dt.year.isin(anos_remover)]
```

```python
data['periodo'] = pd.to_datetime(data['periodo'])
```

```python
data_filtrada.info()
```

```python
data_bienal = data_filtrada.groupby(pd.Grouper(key='periodo', freq='2YE')).mean()
```

```python
data_bienal = data_bienal.dropna(how='all')
data_bienal = data_bienal[~data_bienal.index.year.isin(anos_remover)]
```
